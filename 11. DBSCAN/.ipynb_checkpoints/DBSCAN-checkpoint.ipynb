### THEORY AND INTUITION

```python
# DBSCAN --> density based spatial clustering of applications with noise

# DBSCAN iterates through point and uses two key hyperparameters (epsilon and minimum number of points
# unlike K-Means, it focuses on density as the main factor for cluster assignment of points

# epsilon = distance extended from a point
# min number of points = min number of points in an epsilon distance

# DBSCAN
#     core
#     border
#     outlier

# DBSCAN procedure
# pick a rando point not yet assigned
# determine the point type
# once a core point has been found add all directly reachabke points to the same cluster as core
# repeat until all points have been assigned to a cluster or as an outlier

# DBSCAN is density based while K-Means is distance based
```

```python
def display_categories(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

##### K-MEANS

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
display_categories(model, blobs)
```

```python
model = KMeans(n_clusters=3)
display_categories(model, circles)
```

```python
model = KMeans(n_clusters=3)
display_categories(model, moons)
```

##### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.6)
display_categories(model, blobs)
```

```python
model = DBSCAN(eps=0.15)
display_categories(model, circles)
```

```python
model = DBSCAN(eps=0.15)
display_categories(model, moons)
```

### HYPERPARAMETER THEORY

```python
# increasing epsilon allows more points to be core points which also reults in more border points and less outlier points
# imagine a huge epsilon, all points would be witin the neighborhood and classified as the same cluster
# by changing epsilon we end up choosing how many clusters we are having

# methods of finding an epsilon value
# Run multiple DBSCAN modls varying epsilon and measure:
#     number of clusters
#     number of outliers
#     percentage of outliers

# min number of samples:
# imagine if min number of samples was close to ttal number of points available then very likely all points would become outliers
```

### HYPERPARAMETER TUNING METHODS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
two_blobs = pd.read_csv('cluster_two_blobs.csv')
```

```python
two_blobs_outliers = pd.read_csv('cluster_two_blobs_outliers.csv')
```

```python
sns.scatterplot(data=two_blobs, x='X1', y='X2')
```

```python
sns.scatterplot(data=two_blobs_outliers, x='X1', y='X2')
```

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1')
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN()
```

```python
display_categories(dbscan, two_blobs)
```

```python
display_categories(dbscan, two_blobs_outliers)
```

```python
dbscan = DBSCAN(eps=0.001)
display_categories(dbscan, two_blobs_outliers)
# due tp epsilon being so small dbscan considers all the points to be outliers
```

```python
dbscan = DBSCAN(eps=10)
display_categories(dbscan, two_blobs_outliers)
# due to epsilon being so large dbscan considers all the points to be of one class
```

```python
dbscan = DBSCAN(eps=1)
display_categories(dbscan, two_blobs_outliers)
```

```python
dbscan.labels_
```

```python
# total outliers found
np.sum(dbscan.labels_ == -1)
```

```python
(np.sum(dbscan.labels_ == -1)/len(dbscan.labels_))*100
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001,10,100):
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    perc_outliers = (np.sum(dbscan.labels_ == -1)/len(dbscan.labels_))*100
    outlier_percent.append(perc_outliers)
    
    
```

```python
sns.lineplot(x=np.linspace(0.001,10,100), y=number_of_outliers)
plt.xlim(0,2)
```

```python
sns.lineplot(x=np.linspace(0.001,7,100), y=number_of_outliers)
plt.xlim(0,2)
plt.ylim(0,10)
plt.hlines(y=3, xmin=0,xmax=2,color='red')
```

```python
outlier_percent = []
number_of_outliers = []

for n in np.arange(1,100):
    dbscan = DBSCAN(min_samples=n)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    perc_outliers = (np.sum(dbscan.labels_ == -1)/len(dbscan.labels_))*100
    outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.arange(1,100), y=outlier_percent)
plt.xlabel('Min number of samples')
plt.ylabel('Pct of points classified as an outlier')
```

```python
num_dim = two_blobs_outliers.shape[1]
dbscan = DBSCAN(min_samples=2*num_dim, eps=0.75)
display_categories(dbscan, two_blobs_outliers)
```
