```python
# random forest are known as ensemble learners, since they rely on an ensemble of models(multiple decision trees)
```

```python
# we could try adjusting rules such as
# splitting criterion
# min gini imp decrease
# setting depth limits
# limits on number of terminal leaf node

# however even with all these added hyperparameter adjustments the single decision tree is still limited:
# single feature for root node
# splitting criteria can lead to some features not being used
# potential for overfitting to data
```

```python
# key idea of random forrest:
# create subsets of randomly picked features at each potential split
# the problem is addressed of all feature not included
# we use all the features (may not be on the root node) in splitting

# as there will be alot of trees we simply tally up the majority votes to come on a decision
# eg: tree 1: y=0 , tree 2: y=0, tree 3: y=1 ==> (0,0,1)
# the result will be y==0
# in continuous problem we can average out the results for each tree
```

### KEY HYPERPARAMETERS

```python
# most of the rules or hyperparametrs in the random forrest are the same as decision tree
# there are just a few that are diff:
#    - number of estimators
#          ->how many decision trees to use total on forrest
#    - number of features
#          -> how many features to include in each subset
#    - bootsrap samples
#          -> allow for bootsrap sampling of each training subset of features
#    - out-of-bag error
#          -> calculate OOB error during training
```

### NUMBER OF ESTIMATORS AND FEATURE IN SUBSETS

```python
#  how to choose number of trees:
# reasonable default value: 100
# publications suggest 64-128 trees
# cross validate a grid search of trees
# plot errors versus number of trees (same as KNN elbow method)
```

```python
# after a certain number of trees:
# diff random selections dont reveal any more information
#   -> trees become higly corr
# duplication of trees starts happening

# overfiting is of minimal concern
```

```python
# how many features to include:
# log2(N+1)   --> log base 2 of (N+1) where N is the number of features
# start with sqrt(N) is preffered
```

```python
# hyperparameter review:
# number of estimators:
#     -> start with 100 as default, grid search for higher values
# number of features:
#     -> stat with sqrt(N), grid search for other possible values (N/3)
```

### BOOTSTRAPPING AND OUT OF BAG ERROR

```python
# bootstrapping?
# a term used to describe "random sampling with replacement"
# we allow for using one sampmle more than once that is what with replacement means

# bootstrapping in random forrest
# for each split we are randomly selecting a subset of features
# ths random dubset of features helps create more diverse trees that are not correlated to each other
# to further differentiate trees we could bootstrap a selection of rows fro each split
# this results in two randomized training components:
#     -> subset of feature used
#     -> bootstrapped rows of data
# bootstrapping can be set to false during training(true by default)
# it is an hyperparameter that reduces correlation between trees since trees are then trained on different subsets of feature coluns and data rows
```

```python
# Out of Bag Error

# what is bagging?
# recall for random forrest we use bootstrapped data and then calculate the prediction based on aggregation
#     -> classification: most voted Y class
#     -> Regresion: Average predicted Ys

# in doing bootstrapping there will be some of the rows that wont be used in constructing the trees
# these rows are known as out of bag samples and can be used in testing the model
#  OOB error woukd be the comparision between the ,predicted value and the real values of the OOB samples
# OOB score does not affect the training process
# it is seperate form bootstrapping, OOB score is optional way measuring performance.
# OOB can only be used on the bootstrapped model that has not used all of the OOB samples
# default for OOB is false
```

##### PART 3 POLYNOMIAL REGRESSION MODEL

```python
def run_model(model, X_train, y_train, X_test, y_test):
    
    
#     FIT MODEL TRAINING
    model.fit(X_train, y_train)
#   print(f'BP:{model.best_params_}')

#     GET METRICS
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    mae = mean_absolute_error(y_test, preds)
    print(f'MAE:{mae}')
    print(f'RMSE:{rmse}')


#     PLOT RESULTS MODEL SIGNAL RANGE
    signal_range = np.arange(0,100)
    signal_preds = model.predict(signal_range.reshape(-1,1))
    
    plt.figure(figsize=(12,8), dpi = 200)
    sns.scatterplot(x='Signal', y='Density', data=df, color = 'black')
    
    plt.plot(signal_range,signal_preds)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=6), LinearRegression())
```

```python
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
# while increasing the degrees we will pick too much noise
# what if we have data sets stretching to more than 100hz and we donot know what to expect
# in this case we can go to more robust methods such as K-nearest neighbors or decision trees reg, or support vector reg
```
